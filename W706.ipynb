```python
#important library
import pandas as pd
import glob
```

```python
# loading all csv file names
files = glob.glob("csvs_per_year/*.csv")
```

```python
# reading csv files
for index,file in enumerate(files):
    files[index] = pd.read_csv(file,index_col='date',parse_dates=['date'])
```

```python
# merging all csv files
df = pd.concat(files)
```
